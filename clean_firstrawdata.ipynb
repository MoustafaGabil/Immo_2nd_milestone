```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

 # importing dataset

# Load the file to the data_set
data_set= pd.read_csv('filtered_data_residential.csv')

# Remove commas and periods from the specified column
data_set['Street'] = data_set['Street'].str.replace('[,\.]', '', regex=True)

# Remove white spaces  (leading, trailing, or extra spaces between words)
data_set = data_set.apply(lambda x: x.str.strip().str.replace('\s+', ' ', regex=True) if x.dtype == "object" else x)
# fill None when the value is empty
data_set = data_set.fillna('None')

# Create a new CSV file with cleaned data
data_set.to_csv('test.csv') 


```
